# how runtime can effect in popularity of movies and profit in last 10 years from 2005 to 2015

```python
df["year"]=df["release_date"].dt.year
```

```python
df_s= pd.DataFrame(df[df["year"]>2005])
df_s["year"].unique()
```

```python
#the following two line of code run one time to replace rows contain 0 to NAN then drop NAN rows
#df_s.replace(0, pd.np.nan)
#df_s.dropna(axis=0, how='any', subset="runtime").fillna(0).astype(int)

df_s[df_s["runtime"]==0]#check that there`s no rows equal 0 in runtime column
```

```python
df_s["popularity"].describe()
```

```python
bin_edges=[0.001,0.39,0.70,0.76,33]


bin_names = ['low', 'medium', 'high', 'very_high']
```

```python
df_s['popularity_levels'] = pd.cut(df_s['popularity'], bin_edges, labels=bin_names)
df_s.head()
```

```python
def groupby(col_name):
    x=pd.DataFrame()
    X=df_s.groupby(col_name).mean().runtime
    return X
```

```python
groupby('popularity_levels')
```

```python
groupby('popularity_levels').plot(kind='bar',facecolor='red',figsize=(8,8));
plt.ylabel('runtime')
plt.xlabel('popularity_levels')
plt.title('relation between runtime and popularity_levels')  
```

```python
#this relation between runtime and popularity_levels
#so movies has more than  100 min has very high popularity_levels
```

```python
bin_edges=[1.000000e+00,2.520261e+07,3.352222e+07,5.236377e+07,2.544506e+09]
bin_names = ['low', 'medium', 'high', 'very_high']
```

```python
df_s['profit_levels'] = pd.cut(df_s["profit"], bin_edges, labels=bin_names)

```

```python
x=groupby('profit_levels')
```

```python
x.plot(kind='bar',facecolor='green',figsize=(8,8));
plt.ylabel('runtime')
plt.xlabel('profit_levels')
plt.title('relation between runtime and profit_levels')  
```

```python
#this relation between runtime and profit_levels
#so movies has more than  112 min has very high profit_levels
```
